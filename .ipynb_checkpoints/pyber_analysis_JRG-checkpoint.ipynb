# Pyber Ride Sharing Analysis

This analysis utilizes two fictional ride sharing data sets. One contains the driver count and type of 120 different cities. The second set contains the city, date, fare, and ID of every ride in the company's history. The objective of this analysis is to explore the relationship between city type and average fare, number of rides, and number of drivers per city.

```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import datetime

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_data = pd.read_csv(city_data_to_load)
ride_data = pd.read_csv(ride_data_to_load)
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types

# x-axis:Total Rides per city
# y-axis:Avg Fare per city
# legend:city type
# marker size: driver count per city
tot_rides = pd.DataFrame(ride_share.groupby(by = 'city')['ride_id'].count()).sort_index()

avg_fare = pd.DataFrame(ride_share.groupby(by = 'city')['fare'].mean()).sort_index()

city_types = pd.DataFrame(city_data.set_index('city')['type']).sort_index()

driver_count = pd.DataFrame(city_data.set_index('city')['driver_count']).sort_index()

# # Build the scatter plots for each city types
# plt.scatter([urban_rides,rural_rides], [urban_fare,rural_fare],s = driver_count*2, alpha = 0.5)
fig, ax = plt.subplots()

colors = ['#ff8051','#8bcfed','#ffdf33']
for i, type in enumerate(list(set(city_data['type']))):
    x, y = tot_rides[city_types['type'] == type],avg_fare[city_types['type'] == type]
    scale = driver_count[city_types['type'] == type]
    ax.scatter(x, y, s = scale*10, label = type, alpha = 0.6, c = colors[i], edgecolors = 'black')

# plt.legend(list(set(city_types['type'], )))
# Incorporate the other graph properties

# Create a legend
lgnd = ax.legend(loc = 0, markerscale = .5, title = 'City Type')
# Incorporate a text label regarding circle size
plt.title('Pyber Ride Sharing Data (2016)')
plt.xlabel('Total Number of Rides (Per City)')
plt.ylabel('Average Fare ($)')
plt.grid()
plt.text(43,30,'Note:\nCirclesize correlates with driver count per city.')
lgnd.legendHandles[0]._sizes = [30]
lgnd.legendHandles[1]._sizes = [30]
lgnd.legendHandles[2]._sizes = [30]
# Save Figure
plt.savefig('RideSharingScatter.png')
```

```python
# Show plot
plt.show()
```

## Fares by City Type

The above pie charts display the expected trend that urban cities dominate the business on a total volume basis across all three reported metrics. Similarly, as expected, we see from these pie charts that rural cities account for the smallest piece of the business and employ the least number of drivers. However, the differences between the Total Fares by City Type and Total Drivers by City Type indicate that drivers in rural cities make more money on average than drivers in urban cities. Additionally, comparing the percentages of each city type in the Total Fares by City Type and Total Rides by City Type show that the average fare price is higher in rural cities than in urban cities. Since urban cities account for 80.9% of all of the drivers in the company, we would expect that the urban contribution to revenue to be the same, but urban cities only account for 62.7% of total fares. These visual observations are confirmed by the table below. 

```python
count_by_type = pd.DataFrame(city_data.groupby('type')['type'].count())
fare_by_type['fare/driver'] = fare_by_type['fare']/drivers_by_type['driver_count']
fare_by_type['fare/total rides'] = fare_by_type['fare']/rides_by_type['fare']
fare_by_type['city count'] = count_by_type['type']
fare_by_type['drivers/city'] = drivers_by_type['driver_count']/fare_by_type['city count']
fare_by_type
```

It is expected that there are two main factors that contribute to the higher average fares seen in rural areas. Because people and destinations are geographically mroe spread out in rural areas, it is reasonable to expect that rural trips are longer on average (by time and/or mileage) than urban trips. Also, because there are less drivers in rural areas (~ 4 drivers per rural city vs. ~36 drivers per urban city) there may be more times when the demand for rides becomes larger than the supply of drivers leading to "surge" pricing. Unfortunately, a limitation of this data set is that without knowing the time and distance of each trip (or pricing per mile/per minute of each trip) we are not able to determine how much either of these factors contributes to the fare price.

## Start of operations in each city

The below table contains the date and time of the first ride in each city, sorted in ascending order. It is shown that operations began in most cities in the begining of January of 2018. 8 of the 18 rural cities that are served by Pyber began service on January 27 or later.

```python
ride_share_start = pd.DataFrame(ride_share.groupby('city')['datetime','type'].min())
ride_share_start.sort_values('datetime')
```

```python
ride_share_end = pd.DataFrame(ride_share.groupby('city')['datetime','type'].max())
ride_share_end.sort_values('datetime')
```

```python
ride_share_service_time = pd.merge(ride_share_start, ride_share_end, left_index=True, right_index=True)
ride_share_service_time = ride_share_service_time.rename(columns = {'datetime_x':'Earliest Ride','datetime_y':'Latest Ride','type_x':'Type'})
ride_share_service_time['Time Delta'] = ride_share_service_time['Latest Ride'] - ride_share_service_time['Earliest Ride']
ride_share_service_time[['Earliest Ride','Latest Ride','Time Delta','Type']]
```

```python
service_time_by_type = pd.DataFrame(ride_share_service_time.groupby('Type')['Time Delta'].sum())
service_time_by_type['Average Time'] = service_time_by_type['Time Delta']/count_by_type['type']
service_time_by_type
```

The above table shows that on average Pyber has been operating in urban cities longer than rural cities. This fact coupled with the higher fare per driver in rural cities that is shown in a previous table implies that rural cities are more profitable for drivers than urban cities. Pyber drivers should consider working in rural areas over urban areas if they are reasonable able to do so.
